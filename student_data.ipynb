```python
from os import name
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = pd.read_csv('data.csv')
names = data['Name']
english = data['English']
database = data['Database']
calculus = data['Calculus']
computer_support = data['Computer Support']
ai = data["AI"]
total = data['Total']
```

```python
#barchart visuallization
plt.figure(figsize=(10,6))
barWidth = 0.15

#set position of bar on x axis
r1 = np.arange(len(names))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
```

```python
# make a plot
plt.bar(r1, english, color='blue', width=barWidth, edgecolor='grey', label='english')
plt.bar(r2, database, color='green', width=barWidth, edgecolor='grey', label='Database')
plt.bar(r3, calculus, color='red', width=barWidth, edgecolor='grey', label='Calculus')
plt.bar(r4, computer_support, color='purple', width=barWidth, edgecolor='grey', label='Computer Support')
plt.bar(r5, ai, color='orange', width=barWidth, edgecolor='grey', label='AI')
```

```python
#addlables
plt.xlabel('Names', fontweight='bold', fontsize=15)
plt.ylabel('scores', fontweight='bold', fontsize=15)
plt.xticks([r + barWidth for r in range(len(names))], names ,rotation=45)
plt.title('Marks of Students', fontweight='bold', fontsize=20)
plt.legend()
```

```python
#create legend and show plot
plt.legend(loc='upper left' , bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

```python
#line graph visualization
plt.figure(figsize=(10 , 6))
for i, subject in enumerate([english, database, calculus, computer_support, ai]):
    plt.plot(names, subject, marker='v', label=['English', 'Database', 'Calculus', 'computer support', 'AI'][i])
```

```python
#add labels
plt.xlabel('Names')
plt.ylabel('Scores')
plt.title('Graph of scores by subject')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#pie chart visualization
plt.figure(figsize=(8,8))
plt.pie(total, labels=names, autopct='%1.1f%%', startangle=140, colors=['blue','green','red','purple','orange','yellow'])
plt.title('Total score distibution among students')
plt.show()
```
